14.  Least Frequent Character

```python
s = input("Enter a string: ")

freq = {}

for ch in s:
    freq[ch] = freq.get(ch, 0) + 1

min_freq = min(freq.values())
least_chars = [ch for ch, count in freq.items() if count == min_freq]

print("Least frequent character(s):", least_chars)
print("Frequency:", min_freq)
```

45. Generate possible valid IP addresses from given string

```python
def restore_ip(s):
    res = []
    for i in range(1, len(s)-2):
        for j in range(i+1, len(s)-1):
            for k in range(j+1, len(s)):
                a, b, c, d = s[:i], s[i:j], s[j:k], s[k:]
                if all(map(lambda x: str(int(x)) == x and 0 <= int(x) <= 255, [a, b, c, d])):
                    res.append(f"{a}.{b}.{c}.{d}")
    return res

print(restore_ip(input("Enter digits: ")))
```

3. Consecutive characters frequency

```python
s = input("Enter a string: ")

if not s:
    print("Empty string!")
else:
    prev_char = s[0]
    count = 1

    print("Character : Frequency of consecutive occurrences")

    for char in s[1:]:
        if char == prev_char:
            count += 1
        else:
            print(f"{prev_char} : {count}")
            prev_char = char
            count = 1

    print(f"{prev_char} : {count}")
```

5. Check if a string can become empty by recursive deletion

```python
def reduce_string(s):
    prev = ""
    while s != prev:
        prev = s

        i = 0
        new_s = ""
        while i < len(s):
            if i + 1 < len(s) and s[i] == s[i + 1]:
                # Skip all consecutive duplicates
                dup = s[i]
                while i < len(s) and s[i] == dup:
                    i += 1
            else:
                new_s += s[i]
                i += 1
        s = new_s
    return s

```

6. Minimum number of rotations to obtain a string

```python
def min_rotations(s1, s2):
    if len(s1) != len(s2):
        return -1
    temp = s1 + s1
    if s2 in temp:

        return temp.index(s2)
    return -1

s1 = input("Enter original string: ")
s2 = input("Enter target string: ")

rot = min_rotations(s1, s2)
if rot == -1:
    print("Cannot obtain target string by rotation.")
else:
    print(f"Minimum number of rotations needed: {rot}")
```

23. Remove K length Duplicates

```python
def remove_k_duplicates(s, k):
    stack = []

    for char in s:
        if stack and stack[-1][0] == char:
            stack[-1][1] += 1

            if stack[-1][1] == k:
                stack.pop()
        else:
            stack.append([char, 1])


    return ''.join(char * count for char, count in stack)

s = input("Enter a string: ")
k = int(input("Enter value of K: "))

result = remove_k_duplicates(s, k)
print("String after removing K-length duplicates:", result)
```

4. Maximum Consecutive Substring Occurrence

```python
s = "aaabaaaabaa"
substring = "aa"

max_count = count = 0
i = 0
while i < len(s):
    if s[i:i+len(substring)] == substring:
        count += 1
        i += len(substring)
        max_count = max(max_count, count)
    else:
        count = 0
        i += 1

print(f"Maximum consecutive '{substring}' occurrence: {max_count}")
```

5. Maximum occurring Substring

```python
from collections import Counter

s = "abcababcababc"
k = 2

substrings = [s[i:i+k] for i in range(len(s) - k + 1)]

freq = Counter(substrings)

max_count = max(freq.values())
max_substrings = [sub for sub, count in freq.items() if count == max_count]

print(f"Maximum occurring substring(s) of length {k}: {max_substrings} with count {max_count}")
```

9.  Extract Indices of substring matches

```python
s = "ababcababc"
substring = "ab"

indices = []
start = 0
while start < len(s):
    index = s.find(substring, start)
    if index == -1:
        break
    indices.append(index)
    start = index + 1

print(f"Substring '{substring}' found at indices:", indices)
```

14.  Test substring order

```python
s = "The quick brown fox jumps over the lazy dog"
substrings = ["quick", "fox", "lazy"]

current_index = 0
in_order = True

for sub in substrings:
    index = s.find(sub, current_index)
    if index == -1:
        in_order = False
        break
    current_index = index + len(sub)
print("Substrings appear in order:" if in_order else "Substrings do NOT appear in order.")
```

16. Filter Strings combination of K substrings

```python
from itertools import product

substrings = ["a", "b", "c"]
k = 2

combinations = [''.join(p) for p in product(substrings, repeat=k)]
print("All combinations of length", k, ":", combinations)
```
