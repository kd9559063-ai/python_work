1. Check if string is symmetrical or palindrome

```python
s = input("Enter a string: ")
mid = len(s) // 2

if s[:mid] == s[-mid:]:
    print("Symmetrical")
else:
    print("Not Symmetrical")

if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

8. Capitalize first and last character of each word

```python
s = input("Enter a string: ")
words = s.split()
result = []

for w in words:
    if len(w) > 1:
        w = w[0].upper() + w[1:-1] + w[-1].upper()
    else:
        w = w.upper()
    result.append(w)

print(" ".join(result))
```

14.  Least Frequent Character

```python
s = input("Enter a string: ")

freq = {}

for ch in s:
    freq[ch] = freq.get(ch, 0) + 1

min_freq = min(freq.values())
least_chars = [ch for ch, count in freq.items() if count == min_freq]

print("Least frequent character(s):", least_chars)
print("Frequency:", min_freq)
```

15. Maximum frequency character

```python
s = input("Enter a string: ")

freq = {}

for ch in s:
    freq[ch] = freq.get(ch, 0) + 1

max_freq = max(freq.values())
max_chars = [ch for ch, count in freq.items() if count == max_freq]

print("Maximum frequency character(s):", max_chars)
print("Frequency:", max_freq)
```

16. Odd Frequency Characters

20. Generating random strings until a given string is generated

```python
import random
import string

target = input("Enter the target string: ")
letters = string.ascii_lowercase + string.ascii_uppercase + string.digits + " "
attempt = ""
count = 0

while attempt != target:
    attempt = ''.join(random.choice(letters) for _ in range(len(target)))
    count += 1

print(f"Target string '{target}' generated after {count} attempts!")
```

25. Find all close matches of input string from a list

26. Find uncommon words from two Strings

```python
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

words1 = set(s1.split())
words2 = set(s2.split())

uncommon = words1.symmetric_difference(words2)

print("Uncommon words:", uncommon)
```

28.  Permutation of a string using inbuilt function

29. Check for URL in a String

```python
import re

s = input("Enter a string: ")

pattern = r"(https?://[^\s]+)"

urls = re.findall(pattern, s)

if urls:
    print("URL(s) found in the string:")
    for url in urls:
        print(url)
else:
    print("No URL found in the string.")
```

45. Generate possible valid IP addresses from given string

```python
def restore_ip(s):
    res = []
    for i in range(1, len(s)-2):
        for j in range(i+1, len(s)-1):
            for k in range(j+1, len(s)):
                a, b, c, d = s[:i], s[i:j], s[j:k], s[k:]
                if all(map(lambda x: str(int(x)) == x and 0 <= int(x) <= 255, [a, b, c, d])):
                    res.append(f"{a}.{b}.{c}.{d}")
    return res

print(restore_ip(input("Enter digits: ")))
```
